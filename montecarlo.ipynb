```python
import ltspice 
import numpy as np
import matplotlib.pyplot as plt

#Nombre del archivo .raw
raw_file = "./raws/RRC.raw"

data = ltspice.Ltspice(raw_file)
data.parse()

VLmin = []
ICmax = []

#print('La cantidad de graficos es: %d' %data.getCaseNumber()) #Numero de casos a analizar por el montecarlo


for i in range(0, data.getCaseNumber() - 1):
    nodesVl = data.get_data ('V(vl)', i)
    Vl = np.array(nodesVl)
    VLmin.append(min(Vl))
    plt.plot(Vl, label='Vl')


#Grafico de Vl(t)
plt.xlabel('$Tiempo [ms]$')
plt.ylabel('$V_{L} [V]$')
plt.title('$Analisis$ $Montecarlo$ $de$ $V_{L}$')
plt.grid(True)

plt.show()

#Hacer histograma de los valores minimos de Vl
plt.title('$Histograma$ $de$ $Vl$')#Titulo
plt.xlabel('$Vl_{min}$')#Etiqueta eje x

#Ancho de las barras
bin_width = 0.03

plt.hist(VLmin, width = bin_width , density= True ,bins=20, alpha=0.75)

plt.show()



#Hacer lo mismo pero para IC(C1)

for i in range(0, data.getCaseNumber() - 1):
    nodesIC = data.get_data ('I(C1)', i) *1000
    IC = np.array(nodesIC)
    ICmax.append(max(IC))
    plt.plot(IC, label='IC')
    
#Grafico de Ic(t)
plt.xlabel('$Tiempo [ms]$')
plt.ylabel('$I_{C} [mA]$')
plt.title('$Analisis$ $Montecarlo$ $de$ $I_{C}$')
plt.grid(True)

plt.show()

#Hacer histograma de los valores maximos de IC
plt.title('$Histograma$ $de$ $Ic$')#Titulo
plt.xlabel('$Ic_{max}$')#Etiqueta eje x

#Ancho de las barras
bin_width = 0.08

plt.hist(ICmax, width = bin_width , density= True ,bins=20, alpha=0.75)

plt.show()


```
